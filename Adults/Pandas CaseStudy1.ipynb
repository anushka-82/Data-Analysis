## **17.Replace Salry Values['<=50k and >=50k to 0 and 1**

```python
data['income'].unique()
```

```python
data['income'].value_counts()
```

```python
def salary(sal):
  if sal == '<=50k':
    return 0
  else:
    return 1
```

```python
data['encoded_salary'] = data['income'].apply(salary)
```

```python
data.head(10)
```

```python
data.replace(to_replace=['<=50K', '>50K'], value=[0,1], inplace = True)
data.head(1)
```
